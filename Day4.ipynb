```python
# Function to convert digits in a number to list
#Input 18976
#Output: [1,8,9,7,6]
def NumberListConversion(n):
    li=[]
    while n!=0:
        r=n%10
        li.append(r)
        n=n//10
    li.reverse()
    return li
NumberListConversion(18976)
```

```python
#print factors of given number
def factors(n):
    print("1",end=" ")
    for i in range(2,n):
        if(n%i)==0:
            print(i,end=" ")
    print(n) 
factors(n=int(input("enter number")))
```

```python
#Input:500,520
#Output:500,502,504,506,508,.........520
def AlternateSequence(lb,ub):
    for i in range(lb,ub+2,2):
        print(i,end=" ")
AlternateSequence(500,520)
```

```python
#strong number
def factorial(n):
    f=1
    for i in range(2,n+1):
        f=f*i
    return f
n=input("enter input")
sum=0
for i in n:
    sum=sum+factorial(int(i))
if(sum==int(n)):
    print("strong number")
else:
    print("not strong")
```

```python
def facsum(n):
    li=[]
    f=1
    while(n!=0):
        r=n%10
        for i in range(1,r+1):
            f=f*i
        li.append(f)
        f=1
        n=n//10
    print(sum(li))
facsum(int(input("enter num:")))
```
