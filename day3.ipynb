### sum of even digit

```python
def EvenDigitSum(n):
    s=0
    while n!=0:
        r = n%10
        if r%2==0:
            s=s+r
        n=n//10
    return s
n=int(input("enter the number"))
k=EvenDigitSum(n)
print(k)
```

### Largest Number

```python
def LargeNumber(n):
    large=0
    while n!=0:
        r = n%10
        if large<r:
            large=r
        n=n//10
    return large
L=LargeNumber(19535)
L
```

### Factorial function

```python
def Facto(n):
    f=1
    for i in range(1,n+1,1):
        f=f*i
    print("Factorial:"+str(f))
Facto(int(input("enter the number:")))
```
